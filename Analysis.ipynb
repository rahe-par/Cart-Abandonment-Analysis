```python
# First Analysis: Cart Abandonment by Device
```

```python
device_analysis = (
    df.groupby('device')
      .agg(
          total_sessions=('session_id', 'count'),
          cart_additions=('added_to_cart', 'sum'),
          abandoned_carts=('cart_abandoned', 'sum'),
          purchases=('purchase_completed', 'sum')
      )
)

device_analysis['abandonment_rate'] = (
    device_analysis['abandoned_carts'] /
    device_analysis['cart_additions'] * 100
)

device_analysis = device_analysis.round(2)

device_analysis
```

```python
#visualize it

import matplotlib.pyplot as plt

device_analysis['abandonment_rate'].plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Cart Abandonment Rate by Device")
plt.xlabel("Device")
plt.ylabel("Abandonment Rate (%)")
plt.xticks(rotation=0)

plt.show()
```

```python
# creating a reusable function instead of writing separate code for every category

def abandonment_analysis(df, column):
    analysis = (
        df.groupby(column)
          .agg(
              total_sessions=('session_id', 'count'),
              cart_additions=('added_to_cart', 'sum'),
              abandoned_carts=('cart_abandoned', 'sum'),
              purchases=('purchase_completed', 'sum')
          )
    )

    analysis['abandonment_rate'] = (
        analysis['abandoned_carts'] /
        analysis['cart_additions'] * 100
    ).round(2)

    return analysis.sort_values('abandonment_rate', ascending=False)
```

```python
abandonment_analysis(df, 'device')
abandonment_analysis(df, 'channel')
abandonment_analysis(df, 'region')
abandonment_analysis(df, 'user_type')
abandonment_analysis(df, 'campaign_type')
abandonment_analysis(df, 'weekday')
abandonment_analysis(df, 'month')
```

```python
month_analysis = abandonment_analysis(df, 'month')

month_analysis['abandonment_rate'].plot(
    kind='bar',
    figsize=(8,4)
)

plt.title("Cart Abandonment Rate by Month")
plt.ylabel("Abandonment Rate (%)")
plt.xlabel("Month")
plt.xticks(rotation=45)

plt.show()
```

# FUNNEL DROP ANALYSIS - CONT

```python
# create the funnel metrics table
# Funnel stages
stages = [
    "Sessions",
    "Viewed Product",
    "Added to Cart",
    "Checkout Started",
    "Purchase Completed"
]

users = [
    total_sessions,
    product_views,
    cart_additions,
    checkout_started,
    purchases
]

# Create DataFrame
funnel_df = pd.DataFrame({
    "Stage": stages,
    "Users": users
})

# Calculate conversion from previous stage
funnel_df["Conversion from Previous (%)"] = (
    funnel_df["Users"] / funnel_df["Users"].shift(1) * 100
).round(2)

# First stage is always 100%
funnel_df.loc[0, "Conversion from Previous (%)"] = 100

# Drop-off count
funnel_df["Drop-off"] = (
    funnel_df["Users"].shift(1) - funnel_df["Users"]
)

funnel_df.loc[0, "Drop-off"] = 0

# Drop-off percentage
funnel_df["Drop-off (%)"] = (
    funnel_df["Drop-off"] /
    funnel_df["Users"].shift(1) * 100
).round(2)

funnel_df.loc[0, "Drop-off (%)"] = 0

funnel_df
```

```python
#identify the biggest drop off
largest_drop = funnel_df.loc[funnel_df['Drop-off (%)'].idxmax()]

print("Largest Drop-off Stage")
print("----------------------")
print(f"Stage: {largest_drop['Stage']}")
print(f"Drop-off: {largest_drop['Drop-off (%)']:.2f}%")
```

# BUSINESS INSIGHTS

```python
Executive Insight 1

The e-commerce platform recorded 120,000 customer sessions, but only 6.82% resulted in a 
completed purchase.

Executive Insight 2

The largest customer loss occurred between product viewing and cart addition, where 65.13% of 
users exited the funnel.

Executive Insight 3

Approximately ₹9.47 million worth of carts were abandoned, representing 55.64% of the revenue 
actually generated during the observed period.

Executive Insight 4

Device, marketing channel, month, and user type showed minimal differences in abandonment rates, 
suggesting these factors were not strong drivers of cart abandonment in this dataset.

Executive Insight 5

The dataset exhibits deterministic patterns—for example, all abandoned carts have an order
value of ₹499 and all discounted purchases convert successfully—indicating it is 
synthetically generated. These limitations should be acknowledged when interpreting the results.
```

# RECOMMENDATIONS

```python
| Finding                                           | Recommendation                                                                                                                  |
| ------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------- |
| Large drop between product view and cart addition | Improve product pages with better images, descriptions, customer reviews, pricing clarity, and stronger call-to-action buttons. |
| 40% drop before checkout                          | Simplify the cart experience and reduce unnecessary friction before checkout begins.                                            |
| Nearly 50% drop during checkout                   | Optimize the checkout flow by reducing steps, supporting guest checkout, and offering multiple payment options.                 |
| ₹9.47M in abandoned cart value                    | Implement abandoned-cart recovery campaigns such as reminder emails, push notifications, or limited-time incentives.            |
| Mobile accounts for ~70% of sessions              | Continue prioritizing a smooth and responsive mobile shopping experience.                                                       |
| Synthetic dataset limitations                     | Validate these findings using real transactional data before making strategic business decisions.                               |
```

# CONCLUSION

```python
This project analyzed 120,000 e-commerce sessions to understand customer behavior throughout 
the purchasing funnel. The analysis found that only 6.82% of sessions resulted in a completed 
purchase, with the largest drop-off (65.13%) occurring between product viewing and cart addition.
Customers abandoned carts worth approximately ₹9.47 million, highlighting a significant revenue 
recovery opportunity. While segmentation by device, marketing channel, month, and user type 
showed only minor differences, the funnel analysis identified key stages for optimization. 
Since the dataset is synthetic and contains deterministic relationships, the findings should 
be interpreted as a demonstration of analytical methodology rather than direct evidence of
real-world customer behavior.
```
